```python
Q1=np.percentile(data['Age'],25)
Q2=np.percentile(data['Age'],50)
Q3=np.percentile(data['Age'],75)
```

```python
print(Q1)
print(Q2)
print(Q3)
```

```python
data['Age'].median()
```

```python
IQR=Q3-Q1
IQR
```

```python
lower_limit=Q1-1.5*IQR
up_limit=Q3-1.5*IQR
```

```python
print(lower_limit)
print(up_limit)
```

```python
#finding outliers
outlier=[]
for x in data['Age']:
    if((x>up_limit) or (x<lower_limit)):
        outlier.append(x)
```
